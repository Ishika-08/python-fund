# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot. 

## Bar Plot:
Use: Display the relationship between a categorical variable and a numerical variable.
For example, it can be used to compare the average age of passengers in each class of the Titanic.

## Histogram:
Use: Visualize the distribution of a single continuous variable.
For example, it can be used to show the distribution of ages of passengers on the Titanic.

## Scatter Plot:
Use: Display the relationship between two continuous variables.
For example, it can be used to examine the relationship between age and fare paid by passengers on the Titanic.

## Box Plot:
Use: Summarize the distribution of a continuous variable and compare it across different categories.
For example, it can be used to compare the fares paid by passengers in different classes on the Titanic.

## Heatmap:
Use: Display a 2D data matrix where the values are represented as colors.
For example, it can be used to show correlations between different numerical features of the Titanic dataset.



```python
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Bar Plot
plt.figure(figsize=(8, 6))
sns.barplot(x='class', y='age', data=titanic)
plt.title('Bar Plot of Average Age by Class')
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x='age', kde=True)
plt.title('Histogram of Age Distribution')
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title('Scatter Plot of Age vs Fare')
plt.show()

# Box Plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='fare', data=titanic)
plt.title('Box Plot of Fare by Class')
plt.show()

# Heatmap
plt.figure(figsize=(10, 8))
corr = titanic.corr(numeric_only=True)  # Set numeric_only to True to avoid the FutureWarning
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
# hue for color one color for each entry
# kde=True: Includes a kernel density estimation (KDE) plot along with the histogram to visualize the underlying distribution of the data points.
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', kde=True)
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
# Load the flights dataset from seaborn
flights = sns.load_dataset('flights')

# Reshape the data into a pivot table
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
```
